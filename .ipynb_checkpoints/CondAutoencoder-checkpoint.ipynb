```python
from scipy.sparse import csr_matrix, vstack
from anndata import concat as ad_concat

from anndata import concat as ad_concat

def concatenate_anndata(adata1, adata2):
    # Add batch information
    adata1.obs['batch'] = 0
    adata2.obs['batch'] = 1
    # Concatenate AnnData objects
    concatenated_adata = ad_concat([adata1, adata2], axis=0)
    return concatenated_adata


class Data:
    def __init__(self, x, batch, uce, cell):
        self.x = x
        self.batch = batch
        self.uce = uce
        self.cell = cell

    def to(self, device):
        self.x = self.x.to(device)
        self.batch = self.batch.to(device)
        self.uce = self.uce.to(device)
        self.cell = self.cell.to(device)
        return self

from sklearn.preprocessing import LabelEncoder

def prepare_data_for_training(concat_adata, subset_indices):
    # Convert AnnData object to PyTorch Tensor
    node_features = concat_adata.obsm['X_scGPT'][subset_indices]
    if isinstance(node_features, np.ndarray):
        node_features = torch.FloatTensor(node_features)
    else:  # Assuming it's a sparse matrix
        node_features = torch.FloatTensor(node_features.todense())

    # Additional data
    scgpt_info = torch.from_numpy(concat_adata.obsm['X_scGPT'][subset_indices])
    batch_info = torch.tensor(concat_adata.obs['batch'][subset_indices].values)
    
    # Encode the 'class' column
    label_encoder = LabelEncoder()
    cell_info_encoded = label_encoder.fit_transform(concat_adata.obs['class'][subset_indices].values)
    cell_info = torch.tensor(cell_info_encoded)

    # Create and return a SimpleData object
    return Data(
        x=node_features,
        batch=batch_info,
        uce=scgpt_info,
        cell=cell_info
    )
```

```python
from annoy import AnnoyIndex


def build_annoy_index(embeddings, n_trees=10):
    f = embeddings.shape[1]  # Dimensionality of embeddings
    t = AnnoyIndex(f, 'angular')  # Using 'angular' for cosine similarity
    for i, vec in enumerate(embeddings):
        t.add_item(i, vec)
    t.build(n_trees)
    return t

def generate_triplets_with_pre_filtering(embeddings, batch_labels, num_triplets, knn, cell_types, n_trees=10):
    triplets = []
    valid_indices = np.arange(embeddings.shape[0])

    annoy_indices = {}
    for batch in np.unique(batch_labels):
        batch_embeddings = embeddings[batch_labels == batch]
        annoy_index = build_annoy_index(batch_embeddings, n_trees)
        annoy_indices[batch] = annoy_index

    for _ in range(num_triplets):
        anchor_idx = np.random.choice(valid_indices)
        anchor_embedding = embeddings[anchor_idx:anchor_idx+1]  # Keep the embedding in 2D
        anchor_batch = batch_labels[anchor_idx]

        # Positive examples from opposite batches
        opposite_batches = np.delete(np.unique(batch_labels), np.where(np.unique(batch_labels) == anchor_batch))
        if len(opposite_batches) == 0:
            print("No opposite batch found, skipping this anchor.")
            continue
        opposite_batch = np.random.choice(opposite_batches)
        opposite_annoy_index = annoy_indices[opposite_batch]

        # Query the opposite batch's Annoy index with the anchor's embedding to find KNN
        knn_opposite_indices = opposite_annoy_index.get_nns_by_vector(anchor_embedding[0], knn, include_distances=False)

        # Map local (batch-specific) indices to global indices
        positive_indices = np.array([idx for idx, label in enumerate(batch_labels) if label == opposite_batch])[knn_opposite_indices]

        # Negative examples from the same batch, ensuring not to include the anchor itself
        same_batch_indices = np.array([idx for idx, label in enumerate(batch_labels) if label == anchor_batch and idx != anchor_idx])
        if len(same_batch_indices) < knn:
            print(f"Not enough negative candidates for anchor {anchor_idx}, skipping.")
            continue
        negative_indices = np.random.choice(same_batch_indices, knn, replace=False)

        if len(positive_indices) > 0 and len(negative_indices) > 0:
            positive_idx = np.random.choice(positive_indices)
            #print('positive ', cell_types[positive_idx])
            negative_idx = np.random.choice(negative_indices)
            #print('anchor ',cell_types[anchor_idx])
            triplets.append((anchor_idx, positive_idx, negative_idx))
        else:
            print(f"Skipping triplet generation for anchor {anchor_idx} due to insufficient candidates.")

    return np.array(triplets)
```

```python
import matplotlib.pyplot as plt

def train_autoencoder(model, args, data, lr, epochs, evaluate_interval=10, lamb=0.3):
    optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=0.0001)

    #triplet loss
    triplet_loss_frequency = 50  # Frequency of calculating triplet loss
    
    # Initialize triplet indices to None before the loop
    last_anchor_indices = last_positive_indices = last_negative_indices = None
    
    # Instantiate TripletMarginLoss once outside the loop
    triplet_loss = torch.nn.TripletMarginLoss(margin=1, p=2, reduction='mean').to(device)
    
    data.x = data.x.to(device)
    data.batch = data.batch.to(device)
    data.cell = data.cell.to(device)
    
    # Track the loss at each epoch
    train_losses = []
    
    for epoch in range(epochs):
        model.train()
        optimizer.zero_grad()

        #get latent embeddings
        latent_embeddings = model.encode(data.x, data.batch)

        #reconstruct from latent embeddings
        recon_x = model.decode(latent_embeddings, data.batch)
            
        autoencoder_loss = model.loss(recon_x, data.x)
        
        if epoch % triplet_loss_frequency == 0:

            batch_labels = data.batch.squeeze().cpu().numpy()
            cell_types = data.cell
            embeddings_np = latent_embeddings.detach().cpu().numpy()
            
            triplets = generate_triplets_with_pre_filtering(embeddings_np, batch_labels, 50, 8,cell_types, n_trees=10)
           
            last_anchor_indices = torch.tensor(triplets[:, 0], dtype=torch.long, device=device)
            last_positive_indices = torch.tensor(triplets[:, 1], dtype=torch.long, device=device)
            last_negative_indices = torch.tensor(triplets[:, 2], dtype=torch.long, device=device)

        # Calculate triplet loss using the most recent set of triplets
        anchor = latent_embeddings[last_anchor_indices]
        positive = latent_embeddings[last_positive_indices]
        negative = latent_embeddings[last_negative_indices]
        t_loss = triplet_loss(anchor, positive, negative)

        loss = (1-lamb)*autoencoder_loss + lamb*t_loss
        # print('total_loss: ', loss)
        loss.backward()
        #torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=10)
        optimizer.step()

        train_losses.append(loss.item())
        # Print loss at intervals
        if epoch % evaluate_interval == 0:
            print(f'Epoch {epoch+1}/{epochs}, Loss: {loss.item()}')



    # Plot the loss curve after training
    plt.figure(figsize=(10, 5))
    plt.plot(range(1, epochs+1), train_losses, label='Training Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training Loss Curve')
    plt.legend()
    plt.show()

    return model


def evaluate_final(data, model, device):
    with torch.no_grad():
        # Move tensors to the specified device
        x = data.x.to(device)
        batch = data.batch.to(device)
        
        # Get the latent space embeddings
        embeddings = model.encode(x, batch).cpu().numpy()  # Get the latent embeddings and move to CPU
        
    # Create or update AnnData object
    adata = ad.AnnData(X=embeddings)
    adata.obs['batch'] = data.batch.cpu().numpy()
    adata.obs['class'] = data.cell.cpu().numpy()

    # Convert 'batch' and 'class' columns to categorical
    adata.obs['batch'] = adata.obs['batch'].astype('category')
    adata.obs['class'] = adata.obs['class'].astype('category')

    # Compute UMAP (this might take a while depending on the size of your data)
    sc.pp.neighbors(adata, use_rep='X')  # Compute the neighborhood graph
    sc.tl.leiden(adata, resolution=1.0)
    sc.tl.umap(adata)  # Compute UMAP
    
    # Plot UMAP colored by class and batch
    sc.pl.umap(adata, color=['class'])
    sc.pl.umap(adata, color=['batch'])  # Replace with your label key if you have labels
```

```python
import os
import glob
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from annoy import AnnoyIndex
import matplotlib.pyplot as plt
import scanpy as sc
import anndata as ad
from torch.utils.data import DataLoader, TensorDataset


dir1 = '/work/magroup/ehaber/LLOKI/scgpt/merfish1100'
dir2 = '/work/magroup/ehaber/LLOKI/scgpt/merfish500'



# Assuming you have defined all your classes and functions as in your original code:
# ConditionalAutoencoder, train_autoencoder, evaluate_final, build_annoy_index, generate_triplets_with_pre_filtering, etc.

def iterate_through_directories(zhuang_dir, sec37_dir, device, model_args, lr=0.01, epochs=200):
    # List all files in both directories
    zhuang_files = sorted(glob.glob(os.path.join(zhuang_dir, "*.h5ad")))
    sec37_files = sorted(glob.glob(os.path.join(sec37_dir, "*.h5ad")))

    # Check if both directories have the same number of files
    if len(zhuang_files) != len(sec37_files):
        print("Warning: The number of files in zhuang and sec37 directories do not match.")
        # Handle accordingly, e.g., iterate up to the minimum length
        min_length = min(len(zhuang_files), len(sec37_files))
        zhuang_files = zhuang_files[:min_length]
        sec37_files = sec37_files[:min_length]

    # Initialize the model
    model = ConditionalAutoencoder(
        enc_in_channels=model_args['enc_in_channels'],
        batch_dim=model_args['batch_dim'],
        latent_dim=model_args['latent_dim']
    ).to(device)

    # Initialize optimizer outside the loop for incremental training
    optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=0.0001)
    triplet_loss = torch.nn.TripletMarginLoss(margin=1, p=2, reduction='mean').to(device)

    train_losses = []

    for idx, (z_file, s_file) in enumerate(zip(zhuang_files, sec37_files)):
        print(f"Processing pair {idx+1}/{len(zhuang_files)}:")
        print(f"  Zhuang file: {z_file}")
        print(f"  Sec37 file: {s_file}")

        # Load AnnData objects
        zhuang = ad.read_h5ad(z_file)
        sec37 = ad.read_h5ad(s_file)

        # Assign batch labels
        zhuang.obs['batch'] = 0
        sec37.obs['batch'] = 1

        # Concatenate AnnData objects
        concatenated_subgraph = concatenate_anndata(zhuang, sec37)

        # Prepare data for training
        subset_indices = np.arange(concatenated_subgraph.shape[0])
        data = prepare_data_for_training(concatenated_subgraph, subset_indices)

        # Convert batch labels to tensor and move to device
        data.batch = torch.tensor(data.batch, dtype=torch.float32).unsqueeze(1).to(device)
        data.cell = torch.tensor(data.cell, dtype=torch.long).to(device)  # Ensure cell types are LongTensor

        # Move data to device
        data.x = data.x.to(device)
        data.batch = data.batch.to(device)
        data.cell = data.cell.to(device)

        # Train the model on the current slice pair
        model.train()
        model, epoch_losses = train_autoencoder(
            model=model,
            args=model_args,
            data=data,
            lr=lr,
            epochs=epochs,
            evaluate_interval=50,  # Adjust as needed
            lamb=0.3
        )
        train_losses.extend(epoch_losses)

        # Evaluate the model on the current slice pair
        evaluate_final(data, model, device)

        # Clear CUDA cache to free memory
        torch.cuda.empty_cache()
        current_memory = torch.cuda.memory_allocated(device) / 1e9
        print(f"Current GPU memory usage: {current_memory:.2f} GB\n")

    # After all iterations, plot the aggregated training loss
    plt.figure(figsize=(10, 5))
    plt.plot(range(1, len(train_losses)+1), train_losses, label='Training Loss')
    plt.xlabel('Iteration')
    plt.ylabel('Loss')
    plt.title('Aggregated Training Loss Curve')
    plt.legend()
    plt.show()

    return model

# Example usage:
if __name__ == "__main__":
    # Define directories
    zhuang_dir = "/path/to/zhuang_directory"
    sec37_dir = "/path/to/sec37_directory"

    # Model arguments
    model_args = {
        'enc_in_channels': 512,  # Adjust based on your data
        'batch_dim': 1,           # As per your original code
        'latent_dim': 512
    }

    # Set device
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    # Train the model by iterating through all files
    trained_model = iterate_through_directories(
        zhuang_dir=zhuang_dir,
        sec37_dir=sec37_dir,
        device=device,
        model_args=model_args,
        lr=0.01,
        epochs=200
    )

    # Optionally, save the trained model
    torch.save(trained_model.state_dict(), "trained_conditional_autoencoder.pth")
```
